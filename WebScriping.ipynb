```
import requests #python library 
from bs4 import BeautifulSoup #for web scrape
```

```
#get html content
url = "https://ngodarpan.gov.in/index.php/home/statewise_ngo/14167/27/1" 
response = requests.get(url) 
print(response)
```

```
htmlcontent = response.content
print(htmlcontent)
```

```
soup = BeautifulSoup(htmlcontent, "html.parser")
print(soup.prettify)
```

```
print(soup.findAll('title'))
```

```
print(soup.findAll('div'))
```

```
print(soup.findAll('text'))
```

```
tag = soup.findAll('a')
print(tag)
```

```
anchor = soup.findAll('a')
all_link = set() 
 
for link in anchor:
   if(link.get('href')!='#'):
     linkprint = "https://ngodarpan.gov.in/index.php/home/statewise_ngo/14167/27/1"+link.get('href')
     all_link.add(link) 
     print(linkprint)
```
