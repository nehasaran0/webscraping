```
import bs4
import requests 
import urllib.request
from bs4 import BeautifulSoup
```

```
url= "http://goeduhub.com/" 
response =requests.get(url)
print(response)
```

```
htmlcontent=response.content
#print(htmlcontent)
```

```
soup= BeautifulSoup (htmlcontent, "html.parser")
print(soup)
```

```
print(soup.findAll('title'))
```

```
print(soup.findAll('txt'))
print(soup.findAll('a'))
```

```
anchor = soup.findAll('a')
all_link= set() 
for link in anchor:
    if (link.get('herf')!='#'):
     linkprint="https://www.goeduhub.com/"+link.get('href')
     all_link.add(link)
     print(linkprint)
```
